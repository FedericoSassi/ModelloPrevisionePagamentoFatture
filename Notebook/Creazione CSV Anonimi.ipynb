```python
import pandas as pd

import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df_cerved = pd.read_csv(
    filepath_or_buffer = '../Data/DatiCerved.CSV'
    , delimiter = ';')

df_cerved
```

```python
df_fatture = pd.read_csv(
    filepath_or_buffer = '../Data/ElencoFatture.CSV'
    , delimiter = ','
    , decimal = ','
    , date_parser = True
)

df_fatture
```

```python
Cerved_codCliente = pd.Series(df_cerved.COD_CLIENTE.unique())

Fatture_codiceCliente = pd.Series(df_fatture.CodiceCliente.unique())

ListaCodCliente = Cerved_codCliente.append(Fatture_codiceCliente).unique()

nuovoID = 1

for Codice in ListaCodCliente:
    df_cerved.COD_CLIENTE.replace(
        to_replace = Codice
        , value = nuovoID
        , inplace = True
    )
    
    df_fatture.CodiceCliente.replace(
        to_replace = Codice
        , value = nuovoID
        , inplace = True
    )
    
    nuovoID += 1
    
df_cerved.head(5)
```

```python
df_cerved.to_csv(path_or_buf='../Data/DatiCervedAnonimo.csv',index=False)

df_fatture.to_csv(path_or_buf='../Data/ElencoFattureAnonimo.csv',index=False)
```
